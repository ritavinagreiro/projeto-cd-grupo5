```python
# Histograma das variáveis todas
axes=df.hist(bins= 50,figsize=(20,15))

for ax_row in axes:
    for ax in ax_row:
        if ax is not None:
            coluna = ax.get_title()
            ax.set_title(f"Histograma de {coluna}", fontsize=12)
            ax.set_xlabel("Valor")
            ax.set_ylabel("Frequência")

plt.tight_layout()
plt.show()
```

```python
#Analise das variaveis binarias
binary_vars = ['Telephone', 'Foreign_Worker']

fig, axes = plt.subplots(len(binary_vars), 2, figsize=(10, 4*len(binary_vars)))

for i, var in enumerate(binary_vars):
    
    # Distribuição simples
    sns.countplot(x=var, data=df, color='green', ax=axes[i, 0])
    axes[i, 0].set_title(f'Distribuição - {var.replace("_", " ")}')
    
    # Relação com variável alvo
    sns.countplot(x=var, hue='Creditability', data=df, ax=axes[i, 1])
    axes[i, 1].set_title(f'{var.replace("_", " ")} vs Creditability')

plt.tight_layout()
plt.show()
```

```python
# Histograma das variáveis numéricas 
numerical_vars = [
    'Duration_of_Credit_monthly',
    'Credit_Amount',
    'Duration_in_Current_address',
    'Age_years',
    'Instalment_per_cent',
    'No_of_Credits_at_this_Bank',
    'No_of_dependents'
]

fig, axes = plt.subplots(4, 2, figsize=(15, 16))
axes = axes.flatten()

for i, var in enumerate(numerical_vars):
    sns.histplot(df[var], ax=axes[i], color="violet")
    axes[i].set_title(f'Distribuição de {var}')

for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
import math
n_cols = 2
n_rows = math.ceil(len(numerical_vars) / n_cols)

plt.figure(figsize=(14, 12))

for i, col in enumerate(numerical_vars):
    plt.subplot(n_rows, n_cols, i + 1)
    plt.boxplot(df[col], vert=True)
    plt.title(f'Boxplot - {col}')
    plt.ylabel(col)

plt.tight_layout()
plt.show()
```

```python
categorical_vars = [
    'Account_Balance',
    'Payment_Status_of_Previous_Credit',
    'Purpose',
    'Value_Savings_Stocks',
    'Length_of_current_employment',
    'Sex_Marital_Status',
    'Guarantors',
    'Most_valuable_available_asset',
    'Concurrent_Credits',
    'Type_of_apartment',
    'Occupation'
]

n_cols = 2
n_rows = math.ceil(len(categorical_vars) / n_cols)

plt.figure(figsize=(14, 5 * n_rows))

for i, col in enumerate(categorical_vars):
    plt.subplot(n_rows, n_cols, i + 1)
    
    ax = df[col].value_counts().sort_index().plot(
        kind="bar",
        color="violet",
        edgecolor="black"
    )
    
    plt.title(col.replace("_", " "))
    plt.xticks(rotation=45)
    plt.ylabel("Frequência")
    
    # Valores por cima das barras
    for p in ax.patches:
        ax.text(
            p.get_x() + p.get_width() / 2,
            p.get_height(),
            int(p.get_height()),
            ha='center',
            va='bottom'
        )

plt.tight_layout()
plt.show()
```

# **Outliers**
(Ana Silva)

```python
numerical_vars = [
    'Duration_of_Credit_monthly',
    'Credit_Amount',
    'Duration_in_Current_address',
    'Age_years',
    'Instalment_per_cent',
    'No_of_Credits_at_this_Bank',
    'No_of_dependents'
]

for col in numerical_vars:
    Q1, Q3 = df[col].quantile([0.25, 0.75])
    IQR = Q3 - Q1
    lower, upper = Q1 - 1.5*IQR, Q3 + 1.5*IQR
    
    outliers = ((df[col] < lower) | (df[col] > upper)).sum()
    perc = round(outliers / len(df) * 100, 2)
    
    print(f"{col}: {outliers} outliers ({perc}%)")
```

# **Escalonamento**
(Iara Gomes)

```python
df_before_scaling = df.copy()
```

```python
scale_vars = [
    'Duration_of_Credit_monthly',
    'Credit_Amount',
    'Age_years'
]

scaler = StandardScaler()

df[scale_vars] = scaler.fit_transform(df[scale_vars])
```

# **Novas Variáveis**
(Rita Vinagreiro)

```python
df["Credit_per_Month"] = (df_before_scaling["Credit_Amount"] / df_before_scaling["Duration_of_Credit_monthly"])
df["Credit_Age_Ratio"] = (df_before_scaling["Credit_Amount"] / df_before_scaling["Age_years"])

df[['Credit_per_Month','Credit_Age_Ratio']].head()
```

```python
sns.boxplot(x='Creditability', y='Credit_per_Month', data=df)

plt.title('Credit per Month vs Credit Risk')
plt.xlabel('Creditability (1 = Good Credit, 0 = Bad Credit)')
plt.ylabel('Credit per Month')

plt.show()
```

```python
sns.boxplot(x='Creditability', y='Credit_Age_Ratio', data=df)

plt.title('Credit Age Ratio vs Credit Risk')
plt.xlabel('Creditability')
plt.ylabel('Credit Age Ratio')

plt.show()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

X = df.select_dtypes(include=np.number).drop(columns=['Creditability'])

vif_data = pd.DataFrame()
vif_data["Feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print(vif_data)
```
