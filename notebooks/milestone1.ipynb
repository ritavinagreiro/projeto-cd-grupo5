# Distribuição da variável alvo

(Rita Vinagreiro)

```python
# Contagem da variável alvo
print(df['Creditability'].value_counts())

# Percentagens
print(df['Creditability'].value_counts(normalize=True) * 100)

# Gráfico com cores e legenda
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(
    x='Creditability',
    data=df,
    hue='Creditability',
    palette={0: 'red', 1: 'green'},
    legend=False
)

plt.title('Distribuição da Variável Creditability')
plt.xlabel('Classe')
plt.ylabel('Frequência')

# Legenda lateral manual
plt.legend(
    title='Classe',
    labels=['Mau Crédito (0)', 'Bom Crédito (1)'],
    loc='upper right'
)

plt.show()
```
