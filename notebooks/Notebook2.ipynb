```python
import matplotlib.pyplot as plt

df['Creditability'].value_counts().plot(kind='bar')
plt.title("Distribuição da Variável Alvo - Creditability")
plt.xlabel("Classe")
plt.ylabel("Frequência")
plt.show()
```

```python
# Histograma das variáveis todas
axes=df.hist(bins= 50,figsize=(20,15))

for ax_row in axes:
    for ax in ax_row:
        if ax is not None:
            coluna = ax.get_title()
            ax.set_title(f"Histograma de {coluna}", fontsize=12)
            ax.set_xlabel("Valor")
            ax.set_ylabel("Frequência")

plt.tight_layout()
plt.show()
```

```python
# Matriz de dispersão das variáveis numéricas
df_numericos = df[
    [
        "Duration_of_Credit_monthly",
        "Credit_Amount",
        "Age_years",
        "No_of_dependents"
    ]
]
scatter_matrix(df_numericos, figsize=(12,8), alpha=0.7, diagonal='hist')
plt.suptitle("Matriz de Dispersão das Variáveis Numéricas", fontsize=16)
plt.show()
```

```python
# Histograma das variáveis numéricas 
num_vars =['Duration_of_Credit_monthly','Age_years', 'Duration_in_Current_address', 'Credit_Amount', 'No_of_Credits_at_this_Bank', 'No_of_dependents']
plt.figure(figsize=(12, 20))  

for i, col in enumerate(num_vars):
    plt.subplot(len(num_vars), 2, i+1)  
    df[col].value_counts().plot(kind="bar", color="violet", edgecolor="black")
    plt.title(col)
    plt.xticks(rotation=45)
    plt.ylabel("Frequência")
plt.tight_layout()
plt.show()
```

```python
import math
n_cols = 2
n_rows = math.ceil(len(num_vars) / n_cols)

plt.figure(figsize=(14, 12))

for i, col in enumerate(num_vars):
    plt.subplot(n_rows, n_cols, i + 1)
    plt.boxplot(df[col], vert=True)
    plt.title(f'Boxplot - {col}')
    plt.ylabel(col)

plt.tight_layout()
plt.show()
```

```python
# Histograma das variáveis categóricas 
cat_vars = ["Account_Balance", "Payment_Status_of_Previous_Credit", "Sex_Marital_Status", "Purpose", "Occupation", "Foreign_Worker"]

plt.figure(figsize=(12, 20))  
for i, col in enumerate(cat_vars):
    plt.subplot(len(cat_vars), 2, i + 1)
    
    ax = df[col].value_counts().plot(
        kind="bar",
        color="violet",
        edgecolor="black"
    )
    plt.title(col)
    plt.xticks(rotation=45)
    plt.ylabel("Frequência")
    
    # Adicionar valores por cima das barras
    for p in ax.patches:
        ax.text(
            p.get_x() + p.get_width() / 2,
            p.get_height(),
            int(p.get_height()),
            ha='center',
            va='bottom'
        )

plt.tight_layout()
plt.show()
```
